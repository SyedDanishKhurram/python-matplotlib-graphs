```python
import seaborn as sns
```

```python
sns.barplot(data=df,x='Month',y='Rain',color='blue')
df.plot.bar(x='Month',y=['Rain','Sun','Tmax','Tmin'],color=['green','blue','red','yellow'])
```

```python
df.plot.barh(x='Month',y=['Rain','Sun','Tmax','Tmin'],color=['green','blue','red','yellow'])
```

```python
df.plot.scatter(x='Month',y='Rain',color=['green'],linewidth=10)
```

```python
df.plot(x='Month',y=['Rain','Sun','Tmax','Tmin'],color=['orange','green','purple','black'],linestyle='-.',linewidth=1,marker='*')
```

```python
df.plot(x='Month',y='Sun',color='Red',linestyle='-.',linewidth=4,markerfacecolor='black',markersize=15,marker='*')
plt.xlabel('month of the year')
plt.ylabel('sun value')
plt.title('sun vs month')
plt.show()
```
